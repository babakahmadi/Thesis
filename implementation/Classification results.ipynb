```python
from lib import myDataset
from lib import proposed
from lib import myPlot
from lib import otherMethods
import numpy as np
import time
from lib import runtime

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def barPlot(algs,runtimes,dataName):
    y_pos = np.arange(len(algs))
    plt.bar(y_pos, runtimes,width=.5, color='g', align='center', alpha=0.5)
    plt.xticks(y_pos, algs)
    plt.ylabel('run time')
    plt.title(dataName)
    plt.savefig(dataName+'_bar.png', format='png', dpi=1000)
    plt.show()
```

```python
alg = ['a','b','c','d']
run = [120.12,13,15,11]
barPlot(alg,run,'Dataset')
```

```python
methods = [runtime.myLDA,runtime.myIsomap, runtime.myLLE, 
           runtime.myKPCA, runtime.myICA ]
mine = [runtime.unsupervised, runtime.supervised]
```

```python
X,y = myDataset.loadWine()
runtime = []
names = []
for method in methods:
    runtime.append(method(X,y))
    names.append(method.__name__[2:])
for method in mine:
    runtime.append(method(X,y,5,10)[1])
    names.append(method.__name__[0].upper())
barPlot(names,runtime, "Wine")
```
